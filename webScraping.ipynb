```python
import requests
from bs4 import BeautifulSoup
```

```python
page=requests.get("https://wuzzuf.net/a/IT-Software-Development-Jobs-in-Egypt?ref=browse-jobs-btm").text
soup=BeautifulSoup(page,"html.parser")
```

```python
locationsfulltag=soup.find_all("span", class_="location location-desktop")
```

```python
locationsfulltag
```

```python
locationlist=[]
for i,location in enumerate(locationsfulltag):
  innerlocation=location.span
  locationlist.append(innerlocation.string)
  print(innerlocation.string)
```

```python
locationlist
```

```python
jobtitles=soup.find_all("h2",class_="job-title")
jobtitles
```

```python
jobtitlelist=[]
for j,jobtitleouter in enumerate(jobtitles):
  jobtitle=jobtitleouter.span.string
  jobtitlelist.append(jobtitle)
  print(jobtitle)
jobtitlelist
```

```python
recruitment={'Job title':jobtitlelist,'Location':locationlist}
import pandas as pd  
dataset=pd.DataFrame(recruitment)
```
